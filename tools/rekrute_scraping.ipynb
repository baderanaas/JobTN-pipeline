```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.rekrute.com/offres.html?s=2&p=1&o=1"
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
response = requests.get(url, headers=headers)
response.status_code
```

```python
html = response.content
soup = BeautifulSoup(html, 'html.parser')
```

```python
with open('Rekrute_HTML_Parser.html', 'wb') as file:
    file.write(soup.prettify('utf-8'))
```

```python
from bs4 import BeautifulSoup

divs = soup.find_all("div", {"class": "section"})  # Trouver toutes les sections

for div in divs:
    # Titre du job
    a_tag = div.find("a", class_="titreJob")
    if a_tag:
        job_title = a_tag.get_text().strip()
        job_url = a_tag['href']
        print("Titre du poste:", job_title)
        print("URL:", job_url)
    
    # Description du job (contenu du <span>)
    description_tag = div.find("span", style="color: #5b5b5b;margin-top: 5px;")
    if description_tag:
        job_description = description_tag.get_text().strip()
        print("Description du poste:", job_description)
    
    # Date de publication et date d'expiration
    date_tag = div.find("em", class_="date")
    if date_tag:
        publication_dates = date_tag.find_all("span")
        if len(publication_dates) >= 2:
            start_date = publication_dates[0].get_text().strip()
            end_date = publication_dates[1].get_text().strip()
            print(f"Publication: {start_date} au {end_date}")
        
        # Nombre de postes proposés
        posts_tag = date_tag.find_all("span")[-1]
        if posts_tag:
            num_posts = posts_tag.get_text().strip()
            print("Nombre de postes proposés:", num_posts)
    
    # Informations supplémentaires (secteur, fonction, expérience, etc.)
    info_div = div.find("div", class_="info")
    if info_div:
        ul_tag = info_div.find("ul")
        
        if ul_tag:
            # Afficher l'ensemble de la structure de <ul> et ses <li> pour déboguer
            print("Contenu de <ul> trouvé:")
            for li in ul_tag.find_all("li"):
                text_content = li.get_text().strip()
                print(f"Texte dans <li>: {text_content}")  # Affiche tout le texte des <li>
                
                # Secteur d'activité
                if "Secteur d'activité :" in text_content:
                    sector_links = li.find_all("a")
                    sectors = [sector.get_text().strip() for sector in sector_links]
                    print("Secteurs d'activité:", ", ".join(sectors))
                
                # Fonction
                elif "Fonction :" in text_content:
                    function_links = li.find_all("a")
                    functions = [function.get_text().strip() for function in function_links]
                    print("Fonctions:", ", ".join(functions))
                
                # Expérience requise
                elif "Expérience requise :" in text_content:
                    experience_link = li.find_next("a")
                    if experience_link:
                        experience = experience_link.get_text().strip()
                        print("Expérience requise:", experience)
                
                # Niveau d'étude demandé
                elif "Niveau d'étude demandé :" in text_content:
                    education_link = li.find_next("a")
                    if education_link:
                        education = education_link.get_text().strip()
                        print("Niveau d'étude demandé:", education)
                
                # Type de contrat
                elif "Type de contrat proposé :" in text_content:
                    contract_link = li.find_next("a")
                    if contract_link:
                        contract_type = contract_link.get_text().strip()
                        print("Type de contrat:", contract_type)
                
                # Télétravail
                elif "- Télétravail :" in text_content:
                    telework = text_content.split(":")[-1].strip()
                    print("Télétravail:", telework)
```
