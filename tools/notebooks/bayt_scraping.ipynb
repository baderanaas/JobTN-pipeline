```python
import requests
from bs4 import BeautifulSoup

# URL de la première page
url = "https://www.bayt.com/en/tunisia/jobs/"

def get_jobs(url):
    # Cookies récupérés depuis le navigateur
    cookies = {
        "MSESID0": "4071580746954806558262,0,0,0,OUP16K,0,0,0004a201f5a2758dc4d43397cffb69fb",
        # Ajoutez ici les autres cookies si nécessaire
    }
    
    all_jobs = []  # Liste pour stocker toutes les offres d'emploi

    current_page = 1  # Variable pour suivre la progression

    while True:
        print(f"Scraping page {current_page}...")

        # Effectuer la requête pour la page actuelle
        response = requests.get(url, cookies=cookies)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")

        # Extraire les informations des offres d'emploi
        job_listings = soup.find_all("li", class_="has-pointer-d")

        # Vérifier si des offres ont été trouvées
        if not job_listings:
            print("Aucune offre trouvée sur cette page. Fin du scraping.")
            break

        # Ajouter les offres de cette page à la liste générale
        for job in job_listings:
            # Titre de l'emploi
            title = job.find("h2").text.strip()

            # Lien de l'offre
            link = "https://www.bayt.com" + job.find("a")["href"].strip()

            # Employeur
            company =  job.find("span", class_="t-small")
            company = company.text.strip() if company else "N/A"

            # Localisation
            location = job.find("div", class_="t-mute t-small")
            location = location.text.strip() if location else "N/A"

            # Description
            description = job.find("div", class_="jb-descr")
            description = description.text.strip() if description else "N/A"

            # Date de publication
            date_posted = job.find("span", {"data-automation-id": "job-active-date"})
            date_posted = date_posted.text.strip() if date_posted else "N/A"

            # Salaire
            salary_tag = job.find("dt", class_="jb-label-salary")
            salary = salary_tag.text.strip() if salary_tag else "Not Mentioned"

            # Remote
            remote_tag = job.find("dt", class_="jb-label-remote")
            remote = remote_tag.text.strip() if remote_tag else "Not Remote"

            # Ajouter les informations collectées à la liste
            all_jobs.append({
                "Title": title,
                "Link": link,
                "Company": company,
                "Location": location,
                "Description": description,
                "Date Posted": date_posted,
                "Salary": salary,
                "Remote": remote
            })

        # Vérifier si la page suivante existe
        next_page_button = soup.find("a", class_="jsAjaxLoad", text="More Results")
        if next_page_button:
            # Extraire l'URL de la page suivante et incrémenter le compteur de page
            next_url = "https://www.bayt.com" + next_page_button["href"]
            url = next_url  # Mettre à jour l'URL pour la page suivante
            current_page += 1
        else:
            print("Aucune page suivante trouvée. Fin du scraping.")
            break
    
    return all_jobs

# Appeler la fonction pour démarrer le scraping
jobs = get_jobs(url)

# Afficher les résultats
for job in jobs:
    print(job)
```
