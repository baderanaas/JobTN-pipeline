```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.rekrute.com/offres-emploi-tunisie.html"
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

response = requests.get(url, headers=headers)
response.status_code 
```

```python
html = response.content
soup = BeautifulSoup(html, 'html.parser')
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Trouver toutes les offres d'emploi
    job_listings = soup.find_all('div', class_='col-sm-10 col-xs-12')
    
    for job in job_listings:
        # Extraire le titre du poste et le lien
        title_tag = job.find('a', class_='titreJob')
        job_title = title_tag.text.strip() if title_tag else 'Titre non trouvé'
        
        # Extraction de la description du poste
        job_description_tag = job.find_all('div', class_='info')[1]
        job_description = job_description_tag.find('span').text.strip() if job_description_tag else 'Description du poste non trouvée'
        
        # Extraction de la période de publication
        date_tag = job.find('em', class_='date')
        publication_dates = date_tag.find_all('span') if date_tag else []
        publication_start = publication_dates[0].text.strip() if len(publication_dates) > 0 else 'Date de début non trouvée'
        publication_end = publication_dates[1].text.strip() if len(publication_dates) > 1 else 'Date de fin non trouvée'
        
        # Extraction des détails
        details = {}
        info_tags = job.find_all('li')
        teletravail = 'Non'  # Valeur par défaut
        for tag in info_tags:
            if 'Secteur d\'activité' in tag.text:
                details['Secteur d\'activité'] = tag.text.split(':')[1].strip()
            elif 'Expérience requise' in tag.text:
                details['Expérience requise'] = tag.text.split(':')[1].strip()
            elif 'Niveau d\'étude demandé' in tag.text:
                details['Niveau d\'étude demandé'] = tag.text.split(':')[1].strip()
            elif 'Type de contrat proposé' in tag.text:
                line_content = tag.text.strip()
                if '-' in line_content:  # Si la ligne contient les deux infos
                    contract, telework = line_content.split('-', 1)
                    details['Type de contrat proposé'] = contract.split(':')[1].strip()
                    teletravail = telework.split(':')[1].strip()
                else:  # Sinon, extraire seulement le type de contrat
                    details['Type de contrat proposé'] = line_content.split(':')[1].strip()
        
        # Afficher les informations extraites
        print(f"Poste: {job_title}")
        print(f"Lien: {title_tag['href']}")
        print(f"Description du poste: {job_description}")
        print(f"Période de publication: {publication_start} au {publication_end}")
        print("Détails:")
        for key, value in details.items():
            print(f"{key}: {value}")
        print(f"Télétravail: {teletravail}")

        print("\n" + "="*50 + "\n")
```
